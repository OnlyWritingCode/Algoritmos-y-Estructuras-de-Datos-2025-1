1. Completar la clase `Nodo` y `ListaDoblementeEnlazada` para agregar correctamente cada nodo en ambas direcciones (`siguiente` y `anterior`).
2. Implementar una funci√≥n que permita **imprimir todos los valores desde el primero hasta el √∫ltimo**, y luego desde el √∫ltimo hasta el primero (recorrido en reversa).

````python
class Nodo:
    def __init__(self, valor):
        #DEBEN AGREGAR ALGO
        self.valor = valor
        self.siguiente = None
        self.anterior = None  # completar como parte del ejercicio

class ListaDoblementeEnlazada:
    def __init__(self):
        #DEBEN AGREGAR ALGO
        self.cabeza = None
        self.cola = None  # para poder volver al final

    def agregar(self, valor):
        nuevo = Nodo(valor)
        if self.cabeza is None:
            self.cabeza = nuevo
            self.cola = nuevo
        else:
            #DEBEN AGREGAR ALGO
            self.cola.siguiente = nuevo
            nuevo.anterior = self.cola  # ‚Üê completar enlace hacia atr√°s
            self.cola = nuevo

    def imprimir_ida(self):
        # TODO: implementar impresi√≥n desde cabeza hasta cola
        actual = self.cabeza
        while actual:
            print(actual.valor)
            actual = actual.siguiente

    def imprimir_vuelta(self):
        # TODO: implementar impresi√≥n desde cola hasta cabeza
        actual = self.cola
        while actual:
            print(actual.valor)
            actual = actual.anterior

    #OPCIONAL
    def imprimir_ida_y_vuelta(self):
    # Recorrido desde cabeza a cola
        print("De inicio a fin:")
        actual = self.cabeza
        while actual:
            print(actual.valor)
            actual = actual.siguiente

        print("\nDe fin a inicio:")
        actual = self.cola
        while actual:
            print(actual.valor)
            actual = actual.anterior
````

````python
# Prueba
valores = [23, 45, 654, 1, 34, 54, 213]
lista = ListaDoblementeEnlazada()

for v in valores:
    lista.agregar(v)

print("De inicio a fin:")
lista.imprimir_ida()

print("\nDe fin a inicio:")
lista.imprimir_vuelta()
````
